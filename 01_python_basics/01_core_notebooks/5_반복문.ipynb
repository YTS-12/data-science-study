```python
# 문자열 순회
isUpper = False
isDigit = False
isSpecial = False
isLower = False
password = 'abcD21!'
speciasl_char = "!@#$%"
# 비밀번호 조건 만족여부 확인
for char in password:
    if char.isupper():  #대문자인지 확인
        isUpper = True
    elif char.isdigit():   #숫자인지 확인
        isDigit = True
    if char.islower():
        isLower = True
    if char in speciasl_char:
        isSpecial = True
        check_result = isUpper + isDigit + isLower + isSpecial
print(f"비밀번호 강도 조건 만족여부 : {'만족' if check_result==4  else '불만족'}")
```

```python
# 총일수 % 7 -> calc_day 
#입력일 : 2025-10-5
# 년도 : 1~2024 년도수 * 365 | 366
# 월 : 1~9 월 * 30(4,6,9,11) | 31(1,3,5,7,8,10,12) | 28|29(2-윤년,평년) 
# 일 : 일
```

```python
#날짜 입력받기 > 요일 출력하기
year = int(input('년도를 입력하시오:'))
month = int(input('월을 입력하시오:'))
day = int(input('일을 입력하시오:'))

total_days = 0
for i in range(1,year):
    if i % 400 == 0: 
        total_days += 366
    elif i % 100 == 0:
        total_days += 365
    elif i % 4 == 0:
        total_days += 366
    else:
        total_days += 365
print(f'1st : {total_days}, {year}')

month_days = [0,31,28,31,30,31,30,31,31,30,31,30,31]
for m in range(1,month):
    total_days += month_days[m]
print(f'2nd : {total_days}, {year}')
if month >= 3 : 
    if year % 400 == 0: 
        total_days += 1
    elif year % 100 == 0:
        total_days += 0
    elif year % 4 == 0:
        total_days += 1
    else:
        total_days += 0
print(f'3nd : {total_days}, {year}, {month}')

total_days += (day-1)

print(f'final : {total_days}, {year}, {month}, {day}')

remain = (total_days % 7)
d = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
result = d[remain]
print(f'{year}년 {month}월 {day}일은 {result}입니다.')
```

```python
students = [{'num':'1','name':'kim', 'kor':80, 'math':90, 'eng':100,},
            {'num':'2','name':'park', 'kor':90, 'math':60, 'eng':80,},
            {'num':'3','name':'hong', 'kor':70, 'math':70, 'eng':80,},
            {'num':'4','name':'hong1', 'kor':70, 'math':70, 'eng':80,},
            {'num':'5','name':'hong2', 'kor':70, 'math':70, 'eng':80,}]
```

```python
#학생별 총점, 평균, 순위 -> 속성추가 -> 출력
for s in students:
    s['total'] = s['kor']+s['math']+s['eng']
    s['avg'] = s['total'] / 3
#students
for s in students:
    rank = 1
    for o in students:
        if o ['total'] > s['total']:
            rank += 1
    s['rank']=rank

students
```

```python
#전체 총점, 평균 출력
total = 0
for s in students:
    total += s['total']
t_avg = total/15
total, t_avg
```

```python
#과목별 평균 출력
n = len(students)

kor_sum = 0
math_sum = 0
eng_sum = 0

for s in students:
    kor_sum += s['kor']
    math_sum += s['math']
    eng_sum += s['eng']

print("국어 평균:", f"{kor_sum / n}")
print("수학 평균:", f"{math_sum / n}")
print("영어 평균:", f"{eng_sum / n}")
```

```python
import random
num = random.randint(1,100)
cnt = 0
while True:
    answer = int(input('1에서 100 사이의 숫자를 맞추세요'))
    cnt += 1
    if num == answer:
        print(f'축하합니다. 정답입니다. 도전횟수 : {cnt}')
        break
    elif num > answer:
        print('up')
    else:
        print('down')
```
